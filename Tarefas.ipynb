```python
# Inicializa a lista de tarefas
tarefas = []

def adicionar_tarefa(descricao, categoria, prioridade):
    """Adiciona uma nova tarefa à lista."""
    tarefa = {
        "descricao": descricao,
        "categoria": categoria,
        "prioridade": prioridade
    }
    tarefas.append(tarefa)
    print(f'Tarefa "{descricao}" adicionada com categoria "{categoria}" e prioridade "{prioridade}".')

def listar_tarefas():
    """Lista todas as tarefas."""
    if tarefas:
        print("Tarefas:")
        for i, tarefa in enumerate(tarefas, start=1):
            print(f"{i}. {tarefa['descricao']} | Categoria: {tarefa['categoria']} | Prioridade: {tarefa['prioridade']}")
    else:
        print("Nenhuma tarefa cadastrada.")

def remover_tarefa(indice):
    """Remove uma tarefa da lista pelo índice."""
    if 0 < indice <= len(tarefas):
        tarefa_removida = tarefas.pop(indice - 1)
        print(f'Tarefa "{tarefa_removida["descricao"]}" removida!')
    else:
        print("Índice inválido.")

def salvar_tarefas():
    """Salva as tarefas em um arquivo de texto."""
    with open("tarefas.txt", "w") as arquivo:
        for tarefa in tarefas:
            # Salva as tarefas no formato: descricao;categoria;prioridade
            arquivo.write(f"{tarefa['descricao']};{tarefa['categoria']};{tarefa['prioridade']}\n")
    print("Tarefas salvas no arquivo 'tarefas.txt'.")

def listar_tarefas_salvas():
    """Lista todas as tarefas salvas no arquivo de texto."""
    try:
        with open("tarefas.txt", "r") as arquivo:
            tarefas_salvas = arquivo.readlines()
        if tarefas_salvas:
            print("Tarefas Salvas:")
            for i, linha in enumerate(tarefas_salvas, start=1):
                # Cada linha está no formato: descricao;categoria;prioridade
                descricao, categoria, prioridade = linha.strip().split(";")
                print(f"{i}. {descricao} | Categoria: {categoria} | Prioridade: {prioridade}")
        else:
            print("Nenhuma tarefa salva no arquivo.")
    except FileNotFoundError:
        print("Nenhum arquivo de tarefas salvas encontrado.")

# Menu principal
def main():
    while True:
        print("\nGerenciador de Tarefas")
        print("1. Adicionar Tarefa")
        print("2. Listar Tarefas")
        print("3. Salvar Tarefas")
        print("4. Remover Tarefa")
        print("5. Listar Tarefas Salvas")
        print("6. Sair")
        escolha = input("Escolha uma opção: ")

        if escolha == '1':
            descricao = input("Digite a descrição da tarefa: ")
            categoria = input("Digite a categoria da tarefa (ex.: Trabalho, Pessoal): ")
            prioridade = input("Digite a prioridade da tarefa (Alta, Média, Baixa): ")
            adicionar_tarefa(descricao, categoria, prioridade)
        elif escolha == '2':
            listar_tarefas()
        elif escolha == '3':
            salvar_tarefas()
        elif escolha == '4':
            listar_tarefas()
            try:
                indice = int(input("Digite o número da tarefa a ser removida: "))
                remover_tarefa(indice)
            except ValueError:
                print("Por favor, insira um número válido.")
        elif escolha == '5':
            listar_tarefas_salvas()
        elif escolha == '6':
            print("Saindo...")
            break
        else:
            print("Opção inválida. Tente novamente.")

if __name__ == "__main__":
  main()
```
